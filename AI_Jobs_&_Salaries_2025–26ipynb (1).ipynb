## DataSet(AI Jobs & Salaries 2025–26)
job_category: The broader field or domain of the job within AI and data science.

experience_level: The required experience level for the position (Entry, Mid, Senior, Lead).

years_of_experience: The number of years of professional experience required for the job.

education_required: The minimum educational qualification required for the role.

annual_salary_usd: The estimated average annual salary for the position in USD.

salary_min_usd: The minimum salary offered for the job role.

salary_max_usd: The maximum salary offered for the job role.

city: The city where the job position is located.

country: The country where the job is offered.

remote_work: Indicates whether the job allows remote work or requires on-site presence.

company_size: The size category of the company offering the job (Small, Medium, Large).

industry: The industry sector in which the company operates.

required_skills: The key technical skills required for the job role.

ai_salary_premium_pct: The percentage increase in salary attributed to AI-related expertise.

demand_score: A score representing the market demand level for the job role.

demand_growth_yoy_pct: The year-over-year percentage growth in demand for the role.

benefits_score_10: A rating from 1 to 10 reflecting the quality of job benefits offered.

posting_year: The year when the job posting was published.

posting_month: The month when the job posting was published.

is_senior: A binary indicator showing whether the role is senior-level (1) or not (0).

is_remote_friendly: Indicates whether the company supports remote-friendly work policies.

is_llm_role: Indicates whether the job role involves Large Language Model technologies.

salary_tier: A categorical classification of salary levels (Entry, Mid, Upper-Mid, Senior, Elite

 ## Preprocessing for ML

```python
## Drop Columns
df= data.drop(columns=["job_id",  'annual_salary_usd','salary_min_usd','salary_max_usd','posting_year','posting_month','is_remote_friendly','is_senior','ai_salary_premium_pct','demand_score'])
```

## Feature Engineering

```python
### split  each skills to features and encoding each skill to 0 ,1
skills_encoded = df['required_skills'].str.get_dummies(sep='|')
```

```python
skills_encoded .shape
```

```python
data=pd.concat([data,skills_encoded],axis=1)
data.drop(columns=['required_skills'],inplace=True)
```

```python
### splite for experience_level about year
df['experience_level'] = df['experience_level'].str.replace(')', '', regex=False).str.split(r' \(').str[0]
```

```python
### splite for company_size about number_employeers
df['company_size'] = df['company_size'].str.replace(')', '', regex=False).str.split(r' \(').str[0]
```

### Target Columns

####  benefits_score_10

```python
###   benefits_score_10 to Level Low / Medium / High
df['benefits_score_10']=pd.cut(df['benefits_score_10'],bins=[0,4,7,10],labels=['Low','Medium','High'])
```

```python
## category of benefits_score_10
df['benefits_score_10'].value_counts()
```


####  demand_growth_yoy_pc




```python
## Divide demand_growth_yoy_pct to Level Declining / Stable / Growing
df['demand_growth_yoy_pct']=pd.cut(df['demand_growth_yoy_pct'],bins=[0,33,66,100],labels=['Declining','Stable','Growing'])
```

```python
## category of demand_growth_yoy_pct
df['demand_growth_yoy_pct'].value_counts()
```

### category of salary_tier

```python
## category of salary_tier
df['salary_tier'].value_counts()
```

### Divide data to Features and Target Columns

```python
### split data to features and target
x=df.drop(columns=['salary_tier','benefits_score_10','demand_growth_yoy_pct'])
y=df[['salary_tier','benefits_score_10','demand_growth_yoy_pct']]
```

```python
### category_features
Categorical_column_features=list(x.select_dtypes(include=['object','category']).columns)
Categorical_column_features
```

###  Split Data  To Training Dataset And Testing Dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

### StratifiedKFold

```python
#pip install git+https://github.com/trent-b/iterative-stratification.git
```

```python
from iterstrat.ml_stratifiers import MultilabelStratifiedKFold

mskf = MultilabelStratifiedKFold(n_splits=12, shuffle=True, random_state=42)

for train_index, val_index in mskf.split(x, y):
    x_train, x_val = x.iloc[train_index], x.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]

    # Train multi-output model
```

## Models

#### CatBoost

```python
# Multi-Output Classifier
model = MultiOutputClassifier(
    CatBoostClassifier(
        iterations=300,
        depth=6,
        learning_rate=0.6,
        l2_leaf_reg=0.01,
        loss_function='MultiClass',
        verbose=100,
        early_stopping_rounds=10
    )
)

```

```python
model.fit(x_train, y_train,  cat_features=Categorical_column_features)
```

### Acurracy StratifiedKFold vaildation





```python
from sklearn.metrics import accuracy_score

y_pred6 = model.predict(x_val).squeeze()
```

```python

results = {}
for i, col in enumerate(y_val.columns):
    results[col] = accuracy_score(y_val.iloc[:, i], y_pred6[:, i])
    print(col,":",results[col])
```

```python
## predict_test
y_pred_train=model.predict(x_train).squeeze()
y_pred_test=model.predict(x_test).squeeze()
```

```python
print(accuracy_score(y_train.values.flatten(),y_pred_train.flatten()))
```
